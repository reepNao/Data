```python
employees = {
    'employee_id' : [1,2,3,4],
    'name' : ['Greg', 'Wilson', 'Chase', 'Thirteen'],
}

sales = {
    'employee_id' : [3,4,5,6],
    'total_sales' : [23456, 2512, 2345, 1455]
}
```

```python
employees_df = pd.DataFrame(employees)
sales_df = pd.DataFrame(sales)
```

```python
pd.merge(employees_df, sales_df, on='employee_id', how='left')
```

```python
pd.merge(employees_df, sales_df, on='employee_id', how='right')
```

```python
pd.merge(employees_df, sales_df, on='employee_id', how='outer')
```
