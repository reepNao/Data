# Geometric(p),$E[X]=1/p$




```python
# E[X] = 1/p

p = 0.312
print(1/p)
sMeanX = []
for N in range(1,10000,100):
    sMeanX.append((np.random.geometric(p,N).mean()))
y = np.array(sMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
print(f"value of y is {y[1]}")
```

# Binomial(n,p) $E[X]=np$

```python
# E[X] = np
n = 350
p = 0.312
print(n*p)
sMeanX = []
for N in range(1,10000,100):
    sMeanX.append((np.random.binomial(n,p,N).mean()))
y = np.array(sMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
print(f"value of y is {y[-1]}")
```

# Normal($\mu,\sigma$), $E[X]=\mu$

```python
# E[X] = mu
mu = 2.4
sgma = 3.1

sMeanX = []
for N in range(1,10000,100):
    sMeanX.append((np.random.normal(mu,sgma,N).mean()))
y = np.array(sMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```
